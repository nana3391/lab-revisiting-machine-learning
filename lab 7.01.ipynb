```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('learningSet.csv')
```

```python
print(data.shape)
data.head(10)
```

```python
categorical = data.select_dtypes(object)
categorical.head()
```

```python
categorical.isna().sum()/len(data)
```

```python
categorical =categorical.apply(lambda x: x.replace(" ", np.NaN))
```

```python
nulls_percent_df = pd.DataFrame(categorical.isna().sum()/len(categorical)).reset_index()
nulls_percent_df
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
nulls_percent_df
```

```python
nulls_percent_df[nulls_percent_df['nulls_percentage']!=0]
```

```python
columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>0.85]
columns_above_threshold['column_name']
```

```python
drop_list=[]
```

```python
drop_list = list(columns_above_threshold['column_name'])
print(drop_list)
```

```python
drop_list.extend(['OSOURCE', 'ZIP'])
```

```python
print(drop_list)
```

```python
categorical = categorical.drop(drop_list, axis=1)
```

```python
print(categorical['GENDER'].value_counts())
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
categorical['GENDER'] = categorical['GENDER'].apply(lambda x: x if x in ['M', 'F'] else 'other')
```

```python
categorical.to_csv( 'categoricals123.csv', index= False)
```
